```python
from statsmodels.tsa.stattools import adfuller

df_stationarityTest = adfuller(AMZN["reportedEPS"], autolag='AIC')

print("P-value: ", df_stationarityTest[1])

from statsmodels.graphics.tsaplots import plot_pacf
pacf = plot_pacf(AMZN["reportedEPS"], lags=25)
```

We can observe that stationarity is found as the p-value is < 0.05. The significant spikes occur at lags 1 and 2 so we will use a model of order 2.
